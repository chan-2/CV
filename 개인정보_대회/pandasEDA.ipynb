```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm 
```

```python
disease_map_dict = {
    '1' : '1', # 암
    '2' : '2', # 근골격계
    '3' : '3', # 소화계
    '4' : '3', # 소화계
    '5' : '4', # 내분비계
    '6' : '4', # 내분비계
    '7' : '5', # 심혈관계
    '8' : '5', # 심혈관계
    '9' : '5', # 심혈관계
    '10' : '6', # 호흡계
    '11' : '6', # 호흡계
    '12' : '6', # 호흡계
    '13' : '7', # 기타
    '14' : '7', # 기타
    '15' : '8', # 배설계
    '16' : '2', # 근골격계
    '17' : '2', # 근골격계
    '20' : '7', # 기타 
    # 오탈자 처리
    '133' : '7', # 133 -> 13
    '112' : '6' # 112 -> 12
}
```

```python
data=pd.read_csv("./out.csv")
data.head()
```

```python
for i in tqdm(range(len(data['질병경험'])), total=len(data['질병경험'])):
    tmp_list = data['질병경험'][i].split('|')
    while '' in tmp_list:
        tmp_list.remove('')
    if len(tmp_list) == 0:
        data['질병경험'][i] = '|||||||'
    else:
        new_list = []
        for ele in tmp_list:
            new_list.append(disease_map_dict[ele])
        new_set = set(new_list)
        ori = ['','|','','|','','|','','|','','|','','|','','|','']
        for ele in new_set:
            ori[2*(int(ele)-1)] = ele
        out = ''.join(ori)
        data['질병경험'][i] = out
```

```python
data.to_csv("./dss.csv",index=False)
data.to_excel("./dss.xlsx",index=False)
```
